```python
token_2gam = []
for i in range(len(comment)):
  pop = []
  for tojo in range(len(comment['token'].iloc[i])):
    try:
      pop.append(ast.literal_eval(comment['token'].iloc[i])[tojo]+ast.literal_eval(comment['token'].iloc[i])[tojo+1])
    except:
      pass
  token_2gam.append(pop)
```

```python
token_3gam = []
for i in range(len(comment)):
  pop = []
  for tojo in range(len(comment['token'].iloc[i])):
    try:
      pop.append(ast.literal_eval(comment['token'].iloc[i])[tojo]+ast.literal_eval(comment['token'].iloc[i])[tojo+1]+ast.literal_eval(comment['token'].iloc[i])[tojo+2])
    except:
      pass
  token_3gam.append(pop)
```

```python
comment['token_2g'] = token_2gam
comment['token_3g'] = token_3gam
```

```python
k = []
for i in range(len(comment)):
  L=[]
  for j in Treatment_names:
    for lol in Treatment[j]:
      # Use .iloc to access rows by numerical index
        if lol in ast.literal_eval(comment['token'].iloc[i]):
          L.append(j)
        else:
          L.append('ไม่สามารถระบุได้')
  for j in Treatment_names:
    for lol in Treatment[j]:
      # Use .iloc to access rows by numerical index
        if lol in comment['token_2g'].iloc[i]:
          L.append(j)
        else:
          L.append('ไม่สามารถระบุได้')
  for j in Treatment_names:
    for lol in Treatment[j]:
      # Use .iloc to access rows by numerical index
        if lol in comment['token_3g'].iloc[i]:
          L.append(j)
        else:
          L.append('ไม่สามารถระบุได้')
  L =list(set(L))
  if L != ['ไม่สามารถระบุได้']:
    print(comment['token'].iloc[i])
    print(L)
  if len(L) >= 2:
    L.remove('ไม่สามารถระบุได้')
  k.append(L)
comment['beha'] = k
```
